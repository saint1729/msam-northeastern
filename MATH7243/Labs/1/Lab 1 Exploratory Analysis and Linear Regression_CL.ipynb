# Import Libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Clean Data

```python
#Clean data
ames = pd.read_csv("Amestrain.csv", error_bad_lines=False)
clean = ames['GrLivArea']+ames['BsmtUnfSF']<4000
data = ames[clean]
numberedData = data.select_dtypes(include=['int64','float64'])
CleanData = numberedData.dropna()
```

```python
#get data frames
Names = list(CleanData)
FeaturesData = CleanData.drop(columns='SalePrice')
X = FeaturesData.values
PredictorData = CleanData['SalePrice']
Y = PredictorData.values
ones = np.ones(1100).reshape(-1,1)
X = np.concatenate((ones, X),axis=1)
X
```

```python
#cleaning data based on correlation over 0.4
correlationMatrix = CleanData.corr()
priceCorrelation = correlationMatrix['SalePrice']
filtCorr = priceCorrelation > .4
filtData = priceCorrelation[filtCorr]
```

# Linear Regression

```python
#Linear regression matrix calculation
def Linear_Regression(X, y):
    xT = np.transpose(X)
    yT = np.transpose(y)
    xTx = np.matmul(xT, X)
    xTy = np.matmul(xT, y)
    if np.linalg.det(xTx) == 0:
         raise Exception("Cannot take inverse")
    else:
        xTxinv = np.linalg.inv(xTx)
        theta = np.matmul(xTxinv,xTy)
        return theta
    
    
```

```python
#Calculate Least Squares regression with all features
leastSquaresAll = Linear_Regression(X, Y)
leastSquaresAll
```

```python
#Calculate Least Squares regression with correlated features
leastSquaresCorr = Linear_Regression(Xfilt, Y)
leastSquaresCorr
```

# Predit Function

```python
def createPredictFunction(coefficients, features):
    int_array = coefficients.astype(int)
    for (coefficient, feature) in zip(coefficients,features):
        print(str(coefficient) + "("+feature+")" + " +")
```

```python
# Predict function when including all features
createPredictFunction(leastSquaresAll,Names)
```

```python
# Predict function when including only highly correlated features
createPredictFunction(leastSquaresCorr,FilteredNames)
```

# RSS Cost

```python
def RssCost(X,Y):
    theta = Linear_Regression(X, Y)
    xTheta = np.matmul(X, theta)
    my = xTheta - Y
    norm = np.linalg.norm(my)
    return norm**2
```

```python
# Calculate the cost for all features
RssCost(X,Y)
```

```python
#Calculate the cost for highly correlated features
RssCost(Xfilt,Y)
```
